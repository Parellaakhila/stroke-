# **Preprocessing Techniques**

```python
data.drop('id', axis=1, inplace=True)#drop the id column because it doesn't need the or used in the code because  not the feauture also
print(data.columns)

```

**the dl runs on the numeric data so we run make all them into the numeric data by using the getdummies method**

```python
categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']  #Machine learning/DL models only work with numeric input
data = pd.get_dummies(data, columns=categorical_cols, drop_first=True) #drop_first=True avoids dummy variable trap.
data
```

# **STEP 3: Handle Missing Values with EM (Expectation Maximization)**

```python
# bmi and smoking_status may have missing values.
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer  # fills missing values with nearest values

# Drop 'stroke' and also drop 'age_group' (categorical column used for visualization)
X = data.drop(['stroke', 'age_group'], axis=1)
y = data['stroke']

# Apply imputer
imputer = IterativeImputer(random_state=0)
X_imputed = imputer.fit_transform(X)  # returns a numpy array
X_imputed
```

# **showing the how many missing values are there**

```python
print(np.isnan(X_imputed).sum())  # Total missing values after imputation
```

##**STEP 4: Normalize Features using MinMaxScaler**


```python
#Ensures all values are on the same scale [0, 1] for efficient DNN learning.
from sklearn.preprocessing import MinMaxScaler #(DNNs) perform much better when all input features are roughly on the same scale.

scaler = MinMaxScaler()#MinMaxScaler transforms each feature to a fixed range (usually [0,1]).
X_scaled = scaler.fit_transform(X_imputed)
X_scaled
```

# **STEP 5: Handle Class Imbalance using RandomUnderSampler**

```python
#Only ~1.8% are stroke â†’ model might always predict non-stroke.
#UnderSampling balances both classes.
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42) #every time those were selected 42
X_resampled, y_resampled = rus.fit_resample(X_scaled, y)
X_resampled
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Count the number of samples in each class after undersampling
resampled_counts = pd.Series(y_resampled).value_counts()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(
    resampled_counts.values,
    labels=resampled_counts.index.astype(str),
    autopct='%1.1f%%',
    startangle=90,
    colors=['#4CAF50', '#FF5722']
)
plt.title('Class Distribution After UnderSampling')
plt.show()
```

# **import packages**

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, roc_auc_score, roc_curve
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
```

# **Step 2: Load & preprocess surgical dataset**

```python
# Load surgical data
surgical = pd.read_csv("/content/drive/MyDrive/stroke_prediction/data/Surgical-deepnet.csv")
X_surg = surgical.drop(columns=["complication"])
y_surg = surgical["complication"]

# Impute missing & scale
imp = SimpleImputer(strategy='mean')
X_surg = imp.fit_transform(X_surg)
scaler = StandardScaler()
X_surg = scaler.fit_transform(X_surg)
```

# **Step 3: Build & pre-train model**

```python
pre_model = Sequential([
    Input(shape=(X_surg.shape[1],)),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

pre_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
pre_model.fit(X_surg, y_surg, epochs=30, batch_size=32, verbose=1)
```

**Step 4: Load & preprocess stroke dataset**

```python
stroke = pd.read_csv("/content/drive/MyDrive/stroke_prediction/data/dataset.csv")

# Drop id if exists, one-hot encode categoricals
stroke.drop(columns=["id"], inplace=True, errors='ignore')
stroke = pd.get_dummies(stroke, drop_first=True)

X_stroke = stroke.drop(columns=["stroke"])
y_stroke = stroke["stroke"]

# Impute & scale
X_stroke = imp.fit_transform(X_stroke)
X_stroke = scaler.fit_transform(X_stroke)
```

# **Balance data (SMOTE)**

```python
smote = SMOTE(random_state=42)
X_bal, y_bal = smote.fit_resample(X_stroke, y_stroke)

print("Balanced class counts:", np.bincount(y_bal))
```

```python
kf = KFold(n_splits=10, shuffle=True, random_state=42)
acc_scores, f1_scores, fpr_list, fnr_list, gmean_list = [], [], [], [], []

for fold, (train_idx, test_idx) in enumerate(kf.split(X_bal)):
    print(f"\nðŸ” Fold {fold+1}/10")
    X_train, X_test = X_bal[train_idx], X_bal[test_idx]
    y_train, y_test = y_bal[train_idx], y_bal[test_idx]

    # Build TL-DNN
    model = Sequential([
        Input(shape=(X_train.shape[1],)),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])

    # Transfer weights
    model.layers[4].set_weights(pre_model.layers[4].get_weights())
    model.layers[7].set_weights(pre_model.layers[7].get_weights())

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    callbacks = [
        EarlyStopping(patience=10, restore_best_weights=True),
        ReduceLROnPlateau(patience=5, factor=0.5, verbose=1)
    ]

    model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.2,
              callbacks=callbacks, verbose=1)

    # Predict
    y_pred_prob = model.predict(X_test).ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    fpr = fp / (fp + tn)
    fnr = fn / (fn + tp)
    gmean = np.sqrt((tp / (tp + fn)) * (tn / (tn + fp)))

    acc_scores.append(acc)
    f1_scores.append(f1)
    fpr_list.append(fpr)
    fnr_list.append(fnr)
    gmean_list.append(gmean)
```

```python
print("\n--- 10-Fold Cross Validation Results ---")
print("Avg Accuracy:", round(np.mean(acc_scores), 4))
print("Avg F1 Score:", round(np.mean(f1_scores), 4))
print("Avg FPR     :", round(np.mean(fpr_list), 4))
print("Avg FNR     :", round(np.mean(fnr_list), 4))
print("Avg G-Mean  :", round(np.mean(gmean_list), 4))
```

# **âœ… Stroke Prediction using TL-DNN + SMOTEENN + Class Weights + 10-Fold Cross Validation**

```python
# âœ… Stroke Prediction using TL-DNN + SMOTEENN + Class Weights + 10-Fold Stratified CV (Improved & Stable)

from google.colab import drive
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.utils import class_weight
from sklearn.model_selection import StratifiedKFold
from imblearn.combine import SMOTEENN
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ================================
# STEP 1: Load and preprocess Stroke Dataset
# ================================
drive.mount("/content/drive")
stroke_path = '/content/drive/MyDrive/stroke_prediction/data/dataset.csv'
stroke_data = pd.read_csv(stroke_path)

if 'id' in stroke_data.columns:
    stroke_data.drop('id', axis=1, inplace=True)

categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
stroke_data = pd.get_dummies(stroke_data, columns=categorical_cols, drop_first=True)

X_stroke = stroke_data.drop('stroke', axis=1)
y_stroke = stroke_data['stroke']

imputer = IterativeImputer(random_state=0)
X_stroke_imputed = imputer.fit_transform(X_stroke)

scaler = MinMaxScaler()
X_stroke_scaled = scaler.fit_transform(X_stroke_imputed)

# Balance with SMOTEENN
smote_enn = SMOTEENN(random_state=42)
X_bal, y_bal = smote_enn.fit_resample(X_stroke_scaled, y_stroke)
print("âœ… Balanced class counts:", np.bincount(y_bal))

# ================================
# STEP 2: Pretrain TL-DNN on Transfer Dataset
# ================================
tf_path = '/content/drive/MyDrive/stroke_prediction/data/Surgical-deepnet.csv'
tf_data = pd.read_csv(tf_path)
tf_data = pd.get_dummies(tf_data, drop_first=True)

X_tf = tf_data.drop('complication', axis=1)
y_tf = tf_data['complication']

X_tf_imputed = IterativeImputer().fit_transform(X_tf)
X_tf_scaled = MinMaxScaler().fit_transform(X_tf_imputed)

# Build and pretrain base TL-DNN model
base_model = Sequential([
    Dense(256, activation='relu', input_dim=X_tf_scaled.shape[1]),
    BatchNormalization(),
    Dropout(0.4),
    Dense(128, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),
    Dense(1, activation='sigmoid')
])

base_model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
base_model.fit(X_tf_scaled, y_tf, epochs=30, batch_size=32, verbose=1)

# ================================
# STEP 3: Stratified 10-Fold Cross Validation on stroke data
# ================================
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

acc_list, f1_list, recall_list, precision_list, gmean_list = [], [], [], [], []

for fold, (train_idx, test_idx) in enumerate(skf.split(X_bal, y_bal)):
    print(f"\nðŸ” Fold {fold+1}/10")
    X_train, X_test = X_bal[train_idx], X_bal[test_idx]
    y_train, y_test = y_bal[train_idx], y_bal[test_idx]

    # Compute class weights
    cw = class_weight.compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
    class_weights_dict = {0: cw[0], 1: cw[1]}

    # Build TL-DNN for stroke data (with BatchNorm & lower dropout)
    model = Sequential([
        Dense(256, activation='relu', input_dim=X_train.shape[1]),
        BatchNormalization(),
        Dropout(0.4),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])

    # Transfer pretrained weights from base_model
    for i in range(1, len(model.layers)):
        model.layers[i].set_weights(base_model.layers[i].get_weights())

    model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

    callbacks = [
        EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', patience=4, factor=0.5, verbose=1)
    ]

    history = model.fit(
        X_train, y_train,
        epochs=70,
        batch_size=24,
        validation_split=0.2,
        class_weight=class_weights_dict,
        callbacks=callbacks,
        verbose=1
    )

    # Evaluate
    y_pred_probs = model.predict(X_test).flatten()
    y_pred = (y_pred_probs > 0.5).astype(int)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    gmean = np.sqrt(recall * (cm[0,0] / (cm[0,0]+cm[0,1])))

    print(f"âœ… Fold {fold+1} - Accuracy: {acc:.4f} | F1: {f1:.4f} | Recall: {recall:.4f} | Precision: {precision:.4f} | G-Mean: {gmean:.4f}")

    acc_list.append(acc)
    f1_list.append(f1)
    recall_list.append(recall)
    precision_list.append(precision)
    gmean_list.append(gmean)

# ================================
# STEP 4: Average results
# ================================
print("\nâœ… Final average over 10 folds:")
print(f"Accuracy : {np.mean(acc_list):.4f}")
print(f"F1 Score : {np.mean(f1_list):.4f}")
print(f"Recall   : {np.mean(recall_list):.4f}")
print(f"Precision: {np.mean(precision_list):.4f}")
print(f"G-Mean   : {np.mean(gmean_list):.4f}")
```
